```python
def read_concat_txt(path):
    
    df_results = pd.DataFrame(columns=['frame', 'video_name'])
    
    all_filenames = [i for i in glob.glob('{}*/*.{}'.format(path, extension))]
    file = None
    for file in all_filenames:
        print(file)
        data = pd.read_csv(file, sep=",", header=None)
        data.columns = ["frame", "intensity"]
        
        data['video_name'] = file.split('/')[-2]

        video_name_au = file.replace('.txt', '')
        data['video_name_au'] = video_name_au
        
        data['au'] = video_name_au.split('_au')[-1]

        data['ground truth'] = np.where(data['intensity']== 0, 0, 1)
        
        print(data.shape)        
        df_results = pd.concat([df_results, data], ignore_index=True)
    
    df_results['ground truth'] =df_results['ground truth'].astype(int)
    df_results['intensity'] =df_results['intensity'].astype(int)
        
    return df_results
```

```python
df_results = read_concat_txt(path)
print(df_results.shape)
```
