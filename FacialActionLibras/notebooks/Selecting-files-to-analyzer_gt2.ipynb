```python
import shutil
import glob
from random import randrange
from IPython.display import Image, display
```

```python
def get_all_files_from_directory(path):
    return glob.glob(path)
```

### Selecionando as gravações

```python
path = "../data/raw/*"
```

```python
files = get_all_files_from_directory(path)
files = [f for f in files if f.split('.')[-1] == 'avi']
```

```python
files[:3]
```

```python
c = []
while len(c) < 3:
    _c = randrange(len(files))
    if _c not in c:
        c.append(_c)
```

```python
c
```

```python
len(files)
```

```python
files_selected = list(map(lambda x: files[x], c))
```

```python
files_selected
```

```python
files_selected = ['../data/examples/frames_tamires_gt2/RostoIntensidade-05Quinta-Acalmar/']
```

### Selecionando as imagens / 5 imagens para cada video

```python
f_s_l = []
for _f_selected in files_selected:
    _f_s = get_all_files_from_directory(f"{_f_selected}/*")
    _f_s_l = []
    print(f"{_f_selected}, len: {len(_f_s)}")
    while len(_f_s_l) < 10:
        _f_s_l_ = randrange(len(_f_s))
        if _f_s_l_ not in _f_s_l:
            _f_s_l.append(_f_s_l_)
    fls_selected = list(map(lambda x: f"{_f_s[x]}" , _f_s_l))
    f_s_l.append(fls_selected)
```

```python
f_s_l
```

```python
_r = f_s_l[0][0].split("/")
```

```python
_r[-2]
```

```python
dst = '../data/examples/frames_tamires_gt2/validation_frames/'
for f in f_s_l[0]:
    try:
        shutil.copy(f, dst+f.split('/')[-1])
        print('File', f.split('/')[-1], 'copied successfuly')
    except:
        print("Error occurred while copying file.")
```
